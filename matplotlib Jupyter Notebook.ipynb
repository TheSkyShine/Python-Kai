```python
# Initial plotting of snail data

from matplotlib import pyplot as plt

time = [0, 2, 4, 6, 8]
snail1 = [0, 3, 4.2, 6, 10]

plt.plot(time, snail1)
plt.show()
```

```python
# Let's add some labels and gridlines

plt.plot(time, snail1)

plt.title('Snail Sprinting Stats')
plt.xlabel('Time (minutes)')
plt.ylabel('Distance (inches)')

plt.grid()

plt.show()
```

```python
# How about a race?

snail2 = [0, 5, 4.5, 7.5, 7.5]

# plt.plot(time, snail1) 
# plt.plot(time, snail2)
plt.plot(time, snail1, time, snail2)
plt.title('Snail Sprinting Stats')
plt.xlabel('Time (minutes)')
plt.ylabel('Distance (inches)')
plt.legend(['Snail 1', 'Snail 2'])
plt.grid()

plt.show()
```

```python
# Changing appearances
# Can use worded keyword or symbolic positional arguments (3rd position)

plt.plot(time, snail1, 'k--')
# plt.plot(time, snail1, color= 'black', linestyle= 'dashed')

plt.plot(time, snail2, 'g*')
# plt.scatter(time, snail2, color= 'green')

# plt.plot(time, snail1, 'k-*', time, snail2)

plt.title('Snail Sprinting Stats')
plt.xlabel('Time (minutes)')
plt.ylabel('Distance (inches)')
plt.legend(['Snail 1', 'Snail 2'])
plt.grid()

plt.show()
```

```python
# Axis limits

plt.plot(time, snail1) 
plt.plot(time, snail2)
plt.title('Snail Sprinting Stats')
plt.xlabel('Time (minutes)')
plt.ylabel('Distance (inches)')
plt.legend(['Snail 1', 'Snail 2'])
plt.grid()

plt.xlim(4)
plt.ylim(3, 10)

plt.show()
```

```python
# Separate subplots
# format is (row, column, plot location)

plt.figure(figsize= (15, 5))
# plt.rcParams['axes.grid'] = True

plt.subplot(1,2,1)
plt.plot(time, snail1)
plt.title('Snail 1')
plt.xlabel('Time (minutes)')
plt.ylabel('Distance (inches)')
plt.grid()

plt.subplot(1,2,2)
plt.plot(time, snail2)
plt.title('Snail 2')
plt.xlabel('Time (minutes)')
plt.ylabel('Distance (inches)')
plt.grid()

plt.show()
```

```python
# alternative subplot approach, allows naming of entire figure

fig, (a1, a2) = plt.subplots(1, 2)
fig.suptitle('Snail Sprinting Stats')
fig.set_figheight(5)
fig.set_figwidth(15)

a1.plot(time, snail1,'g-o')
a1.set_title('Snail 1')
a1.set_xlabel('Time (seconds)')
a1.set_ylabel('Distance (inches)')
a1.set_xlim(0, 8.5)
a1.set_ylim(0, 10.5)

a2.plot(time, snail2, 'r:*')
a2.set_title('Snail 2')
a2.set_xlabel('Time (seconds)')
a2.set_ylabel('Distance (inches)')
a2.set_xlim(0, 8.5)
a2.set_ylim(0, 10.5)

plt.show()
```

```python
# Using style sheets and rcParams

import matplotlib

print(matplotlib.style.available)
plt.style.use('Solarize_Light2')
# plt.style.use('default')

# print(matplotlib.rcParams)
matplotlib.rcParams['font.size'] = 5

plt.figure(figsize= (5, 3))
plt.plot(time, snail1, time, snail2)
plt.title('Snail Sprinting Stats')
plt.xlabel('Time (minutes)')
plt.ylabel('Distance (inches)')
plt.legend(['Snail 1', 'Snail 2'])

plt.show()
```
